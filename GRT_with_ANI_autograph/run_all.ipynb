```python
import pandas as pd
import subprocess
import os
import shutil

os.chdir('/work')
print("Current Directory:", os.getcwd())

df = pd.read_csv('dmso_small.csv')

for index, row in df.iterrows():
    print(index, row)
    inp_smiles = row['smiles']
    pKa_EXP = row['exp_pka']
    dielectric_value = row['dielec']
    E_avg_proton = row['E_avg_proton']
    Solvent= row['solvent']
    output_dir = row['id']

    num_conf_rdkit: int = 100
    num_conf_TD: int = 100
    num_conf_gnnis: int = 50
    
    os.makedirs(output_dir, exist_ok=True)
    
    temp_output_notebook_1 = 'pka_pipeline_TD_output.ipynb'
    final_output_notebook_1 = os.path.join(output_dir, temp_output_notebook_1)

    temp_output_notebook_2 = 'full_pipeline_pka_gnnis-1_output.ipynb'
    final_output_notebook_2 = os.path.join(output_dir, temp_output_notebook_2)

    temp_output_notebook_3 = 'full_pipeline_pka_ani_output.ipynb'
    final_output_notebook_3 = os.path.join(output_dir, temp_output_notebook_3)
    
    temp_output_notebook_4 = 'full_pipeline_DFT_output.ipynb'
    final_output_notebook_4 = os.path.join(output_dir, temp_output_notebook_4)
    
    try:
        subprocess.run(
            f"source /opt/conda/etc/profile.d/conda.sh && conda activate torsional_diffusion && "
            f"papermill pka_pipeline_TD.ipynb {temp_output_notebook_1} "
            f"--log-output -p inp_smiles '{inp_smiles}' -p inp_out_path '{output_dir}' "
            f"-p num_conf_TD {num_conf_TD} -k torsional_diffusion",
            shell=True,
            check=True,
            executable="/bin/bash"
        )
        shutil.move(temp_output_notebook_1, final_output_notebook_1)

        subprocess.run(
            f"source /opt/conda/etc/profile.d/conda.sh && conda activate GNNImplicitSolvent && python -m ipykernel install --user --name GNNImplicitSolvent && "
            f"papermill full_pipeline_pka_gnnis-1.ipynb {temp_output_notebook_2} "
            f"--log-output -p inp_smiles '{inp_smiles}' -p output_dir '{output_dir}' "
            f"-p Solvent '{Solvent}' -p num_conf_gnnis {num_conf_gnnis} -k gnnimplicitsolvent",
            shell=True,
            check=True,
            executable="/bin/bash"
        )
        shutil.move(temp_output_notebook_2, final_output_notebook_2)

        # subprocess.run(
        #     f"source /opt/conda/etc/profile.d/conda.sh && "
        #     f"conda activate ani_env && "
        #     f"python -m ipykernel install --user --name=ani_env --display-name='Python (ani_env)' && "
        #     f"papermill full_pipeline_pka_ani.ipynb {temp_output_notebook_3} "
        #     f"--log-output -p inp_smiles '{inp_smiles}' -p output_dir '{output_dir}' "
        #     f"-p pKa_EXP {pKa_EXP} -p E_avg_proton {E_avg_proton} "
        #     f"-p num_conf_rdkit {num_conf_rdkit} -p dielectric_value {dielectric_value} -k ani_env",
        #     shell=True,
        #     check=True,
        #     executable="/bin/bash"
        # )
        # shutil.move(temp_output_notebook_3, final_output_notebook_3)

        subprocess.run(
            f"source /opt/conda/etc/profile.d/conda.sh && "
            f"conda activate ani_env && "
            f"python -m ipykernel install --user --name=ani_env --display-name='Python (ani_env)' && "
            f"papermill full_pipeline_pka_ani_autograph.ipynb {temp_output_notebook_3} "
            f"--log-output -p inp_smiles '{inp_smiles}' -p output_dir '{output_dir}' "
            f"-p pKa_EXP {pKa_EXP} -p E_avg_proton {E_avg_proton} "
            f"-p num_conf_rdkit {num_conf_rdkit} -p dielectric_value {dielectric_value} -k ani_env",
            shell=True,
            check=True,
            executable="/bin/bash"
        )
        shutil.move(temp_output_notebook_3, final_output_notebook_3)


        subprocess.run(
            f"source /opt/conda/etc/profile.d/conda.sh && conda activate base && "
            f"papermill full_pipeline_pka_DFT.ipynb {temp_output_notebook_4} "
            f"--log-output -p inp_smiles '{inp_smiles}' -p output_dir '{output_dir}' "
            f"-p pKa_EXP {pKa_EXP} -p E_avg_proton {E_avg_proton} "
            f"-p num_conf_rdkit {num_conf_rdkit} -p dielectric_value {dielectric_value} -k python3",
            shell=True,
            check=True,
            executable="/bin/bash"
        )
        shutil.move(temp_output_notebook_4, final_output_notebook_4)
        
        
    except subprocess.CalledProcessError as e:
        print(f"❌ Error processing row {index}: {e}")
```
